```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')
ax.plot(x,np.cos(x),label='cos',linestyle='dashed')
ax.set_title('Multiple Lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label');
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashed')
ax.set_title('Multiple Lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend();
#ax.legend(doc=1)
plt.show()
```
